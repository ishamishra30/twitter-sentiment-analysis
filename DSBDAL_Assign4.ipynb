```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/DSBDAL/Facebook_metrics/dataset_Facebook.csv", sep=';')
data
```

```python
data = data.fillna(0)
data.isnull().sum()
```

```python
types = data.groupby("Type")
link = types.get_group("Link")
photo = types.get_group("Photo")
status = types.get_group("Status")
video = types.get_group("Video")
```

```python
data.sort_values(by=['like'])
```

```python
link_melt = link.melt( id_vars=['Type'], value_vars=['like', 'comment'])
link_melt = pd.concat([link_melt[:1],link_melt[-1:]])
link_melt
```

```python
video_melt = video.melt( id_vars=['Type'], value_vars=['like', 'comment'])
video_melt = pd.concat([video_melt[:1],video_melt[-1:]])
video_melt
```

```python
melt = pd.concat([link_melt,video_melt],ignore_index=True)
melt
```

```python
melt.pivot(index='Type', columns='variable')['value']
```
